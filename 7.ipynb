# Failover Simulation

Objective: Switch requests between primary and backup instances when primary fails.

```
import time, random
class Service:
    def __init__(self,name,healthy=True): self.name=name; self.healthy=healthy
    def check(self):
        if random.random()<0.15: self.healthy=False
        return self.healthy

def failover(primary,backup,requests=12):
    for i in range(requests):
        print(f"\nRequest {i+1}:")
        if not primary.check():
            print(f"Primary {primary.name} failed. Switching to {backup.name}")
        else:
            print(f"Primary {primary.name} handled request")
        if not primary.healthy and random.random()<0.3:
            primary.healthy=True; print(f"Primary {primary.name} recovered")
        time.sleep(0.5)

p=Service('primary',True); b=Service('backup',True)
failover(p,b,12)
print("\nFailover simulation done")
```
