# Resource Monitor

Objective: Monitor CPU, memory, and disk usage using psutil.

```
!pip install psutil -q
import psutil, time
def print_stats():
    cpu=psutil.cpu_percent(interval=1)
    mem=psutil.virtual_memory()
    disk=psutil.disk_usage('/')
    print(f"CPU: {cpu}% | Mem: {mem.percent}% ({mem.used//1024**2}MB/{mem.total//1024**2}MB) | Disk: {disk.percent}%")
for i in range(5):
    print_stats()
    time.sleep(1)
print("Monitoring complete")
```
