# SLA Monitor

Objective: Measure response time of service endpoints and check against thresholds.

```
import time, random
try:
    import requests
    R=True
except:
    R=False

ENDPOINTS=['https://httpbin.org/delay/1','https://httpbin.org/delay/2']
THRESHOLD_MS=1500

def check(url):
    if not R:
        latency=random.randint(200,2000)
        status=200 if latency<THRESHOLD_MS else 500
        return status,latency
    import time
    start=time.time()
    try:
        r=requests.get(url,timeout=5)
        latency=int((time.time()-start)*1000)
        return r.status_code,latency
    except:
        return 500,int((time.time()-start)*1000)

for r in range(3):
    print(f"\nRound {r+1}")
    for e in ENDPOINTS:
        status,lat=check(e)
        ok = lat<=THRESHOLD_MS and status==200
        print(f"{e} -> status={status}, latency={lat}ms -> {'OK' if ok else 'VIOLATION'}")
    time.sleep(1)
print("\nSLA monitoring complete")
```
