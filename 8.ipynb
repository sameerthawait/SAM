# Data Replication

Objective: Replicate a data file across multiple simulated storage nodes.

```
import os, shutil
nodes=['node1','node2','node3']
for d in nodes: os.makedirs(d,exist_ok=True)
src='data_file.txt'
if not os.path.exists(src):
    with open(src,'w') as f: f.write('Sample replicated data\n')
for d in nodes:
    dst=os.path.join(d,os.path.basename(src))
    shutil.copy2(src,dst)
    print(f"Replicated to {dst}")
print("Replication complete")
```
