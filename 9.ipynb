# Container Deployment

Objective: Deploy multiple container instances using docker-py or simulate deployment.

```
import time, random
try:
    import docker
    SDK=True
except:
    SDK=False

def simulate_deploy(count=3,image='nginx'):
    for i in range(count):
        print(f"Created container {image}-sim-{i+1}")
        time.sleep(0.4)
    print("Simulation complete")

if SDK:
    try:
        client=docker.from_env()
        for i in range(2):
            img='nginx'
            client.images.pull(img)
            c=client.containers.run(img,detach=True, name=f"{img}-auto-{i+1}", ports={'80/tcp':None})
            print(f"Started {c.name} id={c.short_id}")
        print("Real docker operations attempted")
    except Exception as e:
        print("Docker SDK error:",e)
        simulate_deploy(3)
else:
    simulate_deploy(3)
```
