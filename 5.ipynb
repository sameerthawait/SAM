# Load Balancer

Objective: Simulate load balancing using round-robin and least-connection algorithms.

```
import time, random, threading
class Backend:
    def __init__(self,name): self.name=name; self.active=0
    def handle(self,req):
        self.active+=1
        print(f"{self.name} handling {req} (active={self.active})")
        time.sleep(random.uniform(0.1,0.4))
        self.active-=1

def round_robin(requests, backends):
    idx=0
    for r in requests:
        backends[idx%len(backends)].handle(r); idx+=1

def least_connection(requests, backends):
    for r in requests:
        b=min(backends, key=lambda x:x.active)
        b.handle(r)

bk=[Backend(f"svc-{i+1}") for i in range(3)]
reqs=[f"req-{i+1}" for i in range(12)]
print("Round-robin:")
round_robin(reqs,bk)
time.sleep(0.5)
print("\nLeast-connection:")
least_connection(reqs,bk)
print("\nDone")
```
